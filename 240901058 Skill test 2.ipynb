## Question 1 - Weather Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

weather_data = {
    'Date': ['01-01-2025','02-01-2025','03-01-2025','04-01-2025','05-01-2025',
             '06-01-2025','07-01-2025','08-01-2025','09-01-2025','10-01-2025'],
    'Temperature': [15.2, 16.8, 14.5, 13.0, 17.6, 18.1, 16.2, 12.4, 13.8, 15.5],
    'Humidity': [80, 75, 85, 90, 70, 65, 78, 88, 83, 79],
    'Wind_Speed': [10.5, 8.0, 12.0, 15.2, 7.4, 6.0, 9.1, 14.3, 11.5, 10.0],
    'Rainfall': [0, 0, 1.2, 5.4, 0, 0, 0.5, 2.1, 0, 0]
}
df = pd.DataFrame(weather_data)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# a. Check for missing values and handle them
print("Missing values:\n", df.isnull().sum())
df.fillna(df.mean(numeric_only=True), inplace=True)
print("\nAfter handling missing values:\n", df)
```

```python
# b. Calculate the average temperature for each month
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month')['Temperature'].mean()
print("Monthly Average Temperature:\n", monthly_avg)
```

```python
# c. Visualize the monthly average temperature using a bar plot
plt.figure(figsize=(8,5))
monthly_avg.plot(kind='bar', color='steelblue')
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.tight_layout()
plt.show()
```

## Question 2 - Student Performance Analysis

```python
import pandas as pd
import numpy as np

data = {
    'Student_ID': [101,102,103,104,105,106,107,108,109,110],
    'Name': ['Alice Brown','Ben Carter','Clara Davis','David Evans','Eva Foster',
             'Frank Green','Grace Hall','Henry Ives','Isla Jones','Jack King'],
    'Maths': [85,78,92,65,74,60,88,70,95,82],
    'Physics': [88,75,94,68,78,55,85,72,96,80],
    'Chemistry': [87,80,90,66,76,58,86,74,94,78]
}
df = pd.DataFrame(data)

# a. Summary statistics for each subject
print("Summary Statistics:\n", df[['Maths','Physics','Chemistry']].describe())
```

```python
# b. Calculate the average score for each student
df['Average'] = df[['Maths','Physics','Chemistry']].mean(axis=1)
print("Average score for each student:\n", df[['Name','Average']])
```

```python
# c. Identify students who scored below 60 in more than two subjects
below60 = (df[['Maths','Physics','Chemistry']] < 60).sum(axis=1)
result = df[below60 > 2][['Name']]
print("Students who scored below 60 in more than two subjects:\n", result)
```

## Question 3 - Flight Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
airlines = ['IndiGo','Air India','SpiceJet','Vistara','GoAir']
origins = ['DEL','BOM','MAA','BLR','HYD']
destinations = ['BOM','DEL','BLR','MAA','CCU']

n = 50
flight_data = {
    'Airline': np.random.choice(airlines, n),
    'Origin': np.random.choice(origins, n),
    'Destination': np.random.choice(destinations, n),
    'Departure_Delay': np.random.randint(-10, 120, n)
}
df = pd.DataFrame(flight_data)

# a. Summary statistics for departure delays
print("Summary statistics for departure delays:\n", df['Departure_Delay'].describe())
```

```python
# b. Average delay for each airline
avg_delay = df.groupby('Airline')['Departure_Delay'].mean()
print("Average delay for each airline:\n", avg_delay)
```

```python
# c. Most common departure and arrival destinations
print("Most common departure destination:", df['Origin'].value_counts().idxmax())
print("Most common arrival destination:", df['Destination'].value_counts().idxmax())
```

```python
# d. Box plot of departure delays
plt.figure(figsize=(8,5))
plt.boxplot(df['Departure_Delay'], vert=True, patch_artist=True,
            boxprops=dict(facecolor='steelblue'))
plt.title('Distribution of Departure Delays')
plt.ylabel('Departure Delay (minutes)')
plt.tight_layout()
plt.show()
```

## Question 4 - Euclidean Distance and Dot Product

```python
import numpy as np

# Euclidean Distance for two points in 2D space
p1 = np.array([1, 2])
p2 = np.array([4, 6])
euclidean_dist = np.sqrt(np.sum((p2 - p1)**2))
print("Point 1:", p1)
print("Point 2:", p2)
print("Euclidean Distance:", euclidean_dist)

# Dot Product for two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
dot_product = np.dot(A, B)
print("\nMatrix A:\n", A)
print("Matrix B:\n", B)
print("Dot Product (A . B):\n", dot_product)
```

## Question 5 - Time Series Data: Daily Temperatures

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
dates = pd.date_range(start='2024-01-01', periods=365, freq='D')
temps = 20 + 10 * np.sin(np.linspace(0, 2*np.pi, 365)) + np.random.randn(365) * 2

df = pd.DataFrame({'Date': dates, 'Temperature': temps})
df.set_index('Date', inplace=True)

print("First 5 rows:\n", df.head())
print("\nBasic Statistics:\n", df.describe())
```

```python
monthly_avg = df.resample('ME').mean()
print("Monthly Average Temperature:\n", monthly_avg)
```

```python
rolling_avg = df['Temperature'].rolling(window=7).mean()

plt.figure(figsize=(12,5))
plt.plot(df.index, df['Temperature'], label='Daily Temp', alpha=0.5)
plt.plot(df.index, rolling_avg, label='7-Day Rolling Avg', color='red')
plt.title('Daily Temperature over a Year')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Question 6 - Stock Price Simulation and Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
dates = pd.date_range(start='2024-01-01', periods=365, freq='D')
initial_price = 100
returns = np.random.normal(loc=0.001, scale=0.02, size=365)
prices = initial_price * np.cumprod(1 + returns)

df = pd.DataFrame({'Date': dates, 'Price': prices})
df.set_index('Date', inplace=True)

print("Stock Price Statistics:\n", df.describe())
```

```python
df['MA_7'] = df['Price'].rolling(window=7).mean()
df['MA_30'] = df['Price'].rolling(window=30).mean()

print("First 10 rows with Moving Averages:\n", df.head(10))
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Price'], label='Stock Price', alpha=0.7)
plt.plot(df.index, df['MA_7'], label='7-Day MA', color='orange')
plt.plot(df.index, df['MA_30'], label='30-Day MA', color='red')
plt.title('Daily Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.tight_layout()
plt.show()
```
