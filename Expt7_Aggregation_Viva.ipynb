# Experiment 7 - Data Aggregation and Grouping

**Q1. What is data aggregation in Pandas?**

Data aggregation is the process of combining multiple values into a single summary value using functions like sum, mean, count, min, or max. In pandas, it is performed using functions like `agg()`, `sum()`, and `mean()`.

---

**Q2. What is the purpose of the agg() function in Pandas?**

`agg()` (or `aggregate()`) applies one or more aggregation functions to a DataFrame or Series. It can apply different functions to different columns simultaneously, for example: `df.agg({'A': 'sum', 'B': 'mean'})`.

---

**Q3. What is grouping in data analysis?**

Grouping means splitting data into subsets based on one or more categorical columns. It allows you to analyze each group separately, for example grouping sales data by region to calculate the total sales per region.

---

**Q4. What does the groupby() function do in Pandas?**

`groupby()` splits the DataFrame into groups based on a specified column, applies a function to each group, and combines the results. For example: `df.groupby('Category')['Sales'].sum()` gives total sales per category.

---

**Q5. What is the difference between aggregation and grouping?**

Grouping refers to splitting data into groups based on a column, while aggregation refers to applying a summary function (sum, mean, etc.) to each group. Grouping is usually followed by aggregation to get meaningful results.

---

**Q6. Name some common aggregation functions used in Pandas.**

Common aggregation functions in pandas include `sum()`, `mean()`, `median()`, `min()`, `max()`, `count()`, `std()` (standard deviation), and `var()` (variance). These can be applied directly or through `agg()`.

---

**Q7. How does Pandas handle missing values (NaN) during aggregation?**

By default, pandas skips NaN values during aggregation. For example, `sum()` ignores NaN and adds only valid values. You can change this behavior using `skipna=False` to include NaN in the computation.

---

**Q8. Can we apply multiple aggregation functions to a single column? If yes, how?**

Yes, you can apply multiple aggregation functions using `agg()`. For example: `df['Sales'].agg(['sum', 'mean', 'max'])` returns all three statistics for the Sales column in a single call.

---

**Q9. How can you perform aggregation after grouping data in Pandas?**

After grouping, you chain an aggregation function: `df.groupby('Category')['Sales'].agg(['sum', 'mean'])`. This groups by Category and computes both the sum and mean of Sales for each group.

---

**Q10. What is the output structure when using groupby() with aggregation functions?**

The output is a DataFrame or Series where the group keys become the index and the aggregated values form the columns. For example, grouping by 'Month' and summing 'Sales' gives a DataFrame with Month as index and sum as values.
