## Aim

To implement unsupervised learning-clustering using any predefined dataset

## Description

- Unsupervised learning don't have any target/outcome variable to predict/estimate.

- It is used for clustering population in different groups, which is widely used for segmenting customers in different groups

  - Ex. Apriori algorithm, K-means

```python
# Importing Modules
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# Loading dataset
iris_df = datasets.load_iris()
```

```python
# Available methods on dataset
print("Methods:\n",dir(iris_df))
```

```python
# Features
print("\nFeatures:\n",iris_df.feature_names)
```

```python
# Targets
print("\nTargets:\n",iris_df.target)
```

```python
# Target Names
print("\nTarget names:\n",iris_df.target_names)
label = {0: 'red', 1: 'blue', 2: 'green'}
```

```python
# Dataset Slicing
x_axis = iris_df.data[:, 0]  # Sepal Length
y_axis = iris_df.data[:, 2]  # Sepal Width
```

```python
# Plotting
plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.show()
```
