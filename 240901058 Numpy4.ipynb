```python
import numpy as np
arrl=[10,20,30,40,50] # Corrected to have 5 elements
arr2=[2,4,5,8,10]
a=np.array(arrl)
b=np.array(arr2)
print("Original arrays")
print(a)
print(b)
print("\nVector addition")
print(a+b)
print("\nVector subtraction")
print(a-b)
print("\nVector multiplication")
print(a*b)
print("\nVector division")
print(a/b)
print("\nVector Dot product")
print(a.dot(b))
print("\nScalar multiplication")
sclr=5
print(" scalar value=",sclr)
print("array=",a)
print(" result=",a*sclr)
#Numpy. Vectorize method
def my_func(x, y):
    #"Return x-v if x>v. of otherwise return x+y"
    if x > y:
        return x - y
    else:
        return x + y
print ("\nNumpy.Vectorize method")
print("(Return x-y if x>y, otherwise return x+y)")
arr1_vec=[10,4,20] # Renamed to avoid confusion with arrl
arr2_vec=[2,3,30] # Renamed to avoid confusion with arr2
vec_func=np.vectorize(my_func)
print("array1",arr1_vec)
print("array2:",arr2_vec)
vec_func = np.vectorize(my_func)
print("array1:",arr1_vec)
print("array2:",arr2_vec)
print("result:",vec_func(arr1_vec,arr2_vec))
```
