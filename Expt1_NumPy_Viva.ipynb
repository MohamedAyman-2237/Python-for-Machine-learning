# Experiment 1 - NumPy Viva Questions

**Q1. What is NumPy and its main purpose?**

NumPy (Numerical Python) is a Python library used for numerical computing. Its main purpose is to provide support for large multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on them efficiently.

---

**Q2. How do you create a NumPy array from a Python list?**

You can create a NumPy array from a Python list using the `np.array()` function. For example: `import numpy as np; arr = np.array([1, 2, 3])`. This converts the list into a NumPy array.

---

**Q3. What is the difference between a Python list and a NumPy array?**

A Python list can store elements of different data types and is slower for numerical operations. A NumPy array stores elements of the same data type, is faster due to vectorized operations, and uses less memory than a Python list.

---

**Q4. What does the shape attribute of a NumPy array represent?**

The `shape` attribute of a NumPy array returns a tuple representing the dimensions of the array. For example, a 2D array with 3 rows and 4 columns will have `shape = (3, 4)`.

---

**Q5. What is vectorized computation in NumPy?**

Vectorized computation in NumPy means performing operations on entire arrays without using explicit loops. NumPy applies the operation element-wise internally using optimized C code, making it much faster than Python loops.

---

**Q6. What is broadcasting in NumPy?**

Broadcasting is a mechanism in NumPy that allows arrays of different shapes to be used together in arithmetic operations. NumPy automatically expands the smaller array to match the shape of the larger one without copying data.

---

**Q7. How can you find the mean and standard deviation of a NumPy array?**

You can use `np.mean(arr)` to find the mean and `np.std(arr)` to find the standard deviation of a NumPy array. For example: `np.mean([1,2,3,4])` returns `2.5`.

---

**Q8. How do you perform element-wise addition of two NumPy arrays?**

Element-wise addition is done using the `+` operator or `np.add()` function. For example: `a = np.array([1,2,3]); b = np.array([4,5,6]); result = a + b` gives `[5, 7, 9]`.

---

**Q9. How can you create a NumPy array of zeros with shape (3, 2)?**

You can use `np.zeros((3, 2))` to create an array of zeros with shape (3, 2). This creates a 2D array with 3 rows and 2 columns, all filled with 0.0.

---

**Q10. How do you create an array of evenly spaced numbers between 0 and 10 with a step of 2?**

You can use `np.arange(0, 11, 2)` which returns `[0, 2, 4, 6, 8, 10]`. Alternatively, `np.linspace(0, 10, 6)` creates 6 evenly spaced values between 0 and 10.
