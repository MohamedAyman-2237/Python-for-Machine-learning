# Experiment 3 - Data Loading, Storage & File Formats

**Q1. Name a few common file formats used to store data in data analysis.**

Common file formats include CSV (Comma-Separated Values), Excel (.xlsx), JSON (JavaScript Object Notation), SQL databases, HDF5, and Parquet. Each format has different use cases based on data size and structure.

---

**Q2. What is the difference between CSV and Excel file formats?**

CSV is a plain text format that stores data separated by commas and is lightweight and universally compatible. Excel (.xlsx) is a binary format that supports multiple sheets, formatting, formulas, and charts, but requires specific libraries to read.

---

**Q3. How do you load a CSV file in Python using pandas?**

You can load a CSV file using `pd.read_csv('file.csv')`. This reads the file into a DataFrame. You can specify parameters like `sep`, `header`, `usecols`, and `index_col` for more control.

---

**Q4. What is a DataFrame in pandas?**

A DataFrame is a 2-dimensional labeled data structure in pandas, similar to a spreadsheet or SQL table. It consists of rows and columns where each column can hold a different data type.

---

**Q5. What is the purpose of the read_csv() function?**

The `read_csv()` function reads a CSV file and converts it into a pandas DataFrame. It automatically detects column names from the header row and supports many options for parsing different CSV formats.

---

**Q6. How is data stored internally in a pandas DataFrame?**

Internally, a pandas DataFrame stores data as a collection of NumPy arrays, one per column. Each column is stored as a one-dimensional array with a shared index, enabling fast vectorized operations.

---

**Q7. What is the use of JSON file format?**

JSON (JavaScript Object Notation) is a lightweight, human-readable format used to store and exchange structured data. It is widely used in web APIs and can be read into a pandas DataFrame using `pd.read_json()`.

---

**Q8. What is the difference between text files and binary files?**

Text files store data as human-readable characters (like CSV or JSON), while binary files store data in a binary format (like Excel or HDF5). Binary files are more compact and faster to read but are not human-readable.

---

**Q9. Why is proper data storage important in engineering applications?**

Proper data storage ensures data integrity, easy retrieval, and efficient processing. In engineering, storing data correctly allows reproducibility of experiments, enables future analysis, and avoids data loss or corruption.
