# Experiment 9 - Supervised Learning

**Q1. What is supervised learning?**

Supervised learning is a type of machine learning where the model is trained on labeled data, meaning each input has a corresponding correct output. The model learns the mapping from inputs to outputs and uses it to predict on new data.

---

**Q2. What are input and output variables?**

Input variables (also called features or independent variables) are the data used to make predictions. Output variables (also called labels or dependent variables) are the results the model is trying to predict.

---

**Q3. Give one example of supervised learning.**

Predicting house prices based on features like area, location, and number of rooms is an example of supervised learning (regression). Email spam detection (classifying emails as spam or not) is another example (classification).

---

**Q4. What is a label in supervised learning?**

A label is the correct output or answer associated with each training example in supervised learning. For example, in a dataset of emails, the label would be 'Spam' or 'Not Spam' for each email.

---

**Q5. What is classification?**

Classification is a type of supervised learning where the output is a discrete category or class. For example, predicting whether a tumor is malignant or benign, or classifying an email as spam or not spam.

---

**Q6. What is regression?**

Regression is a type of supervised learning where the output is a continuous numerical value. For example, predicting the price of a house or the temperature tomorrow based on input features.

---

**Q7. Name one supervised learning algorithm.**

K-Nearest Neighbors (KNN) is a simple supervised learning algorithm that classifies a data point based on the majority class of its K nearest neighbors. Other examples include Linear Regression, Decision Tree, and SVM.

---

**Q8. What is training data?**

Training data is the portion of the dataset used to train the machine learning model. The model learns patterns and relationships from this data. Typically, 70-80% of the dataset is used for training.

---

**Q9. What is testing data?**

Testing data is the portion of the dataset used to evaluate the model's performance after training. It is data the model has not seen before, used to check how well the model generalizes to new inputs.

---

**Q10. Why do we split data into training and testing sets?**

We split data to evaluate the model's ability to generalize to unseen data. If we trained and tested on the same data, the model might memorize the data (overfitting) and perform poorly on new real-world inputs.
