Aim:

To perform Data Aggregation and grouping functions

Note:

Aggregation: Aggregation means applying statistical operations like sum, mean, min, max, count
on data. It is used to summarize the data by performing calculations. It uses the function agg().

Grouping: Grouping means splitting the data into groups based on a column or condition. It is
used organize data into categories before performing operations. It uses the function groupby().

Description:
1. Create a DataFrame.
2. Implement Data Aggregation and Grouping functions

```python
import pandas as pd
import numpy as np
df= pd.DataFrame({'A': ['sai', 'sai', 'subu', 'yuvi', 'rithi', 'subu'],'B': ['one', 'one', 'three', 'one', 'three', 'three'],'C': np.random.randint(5, size=6),'D': np.random.randint(5, size=6),'E': np.random.randint(5, size=6)})
print("\nOriginal DataFrame\n",df)
```

```python
#Group by (multi-column): with one column sum
df1=df.groupby(['A', 'B'], as_index=False)['C'].sum()
print("\nOriginal DataFrame\n",df)
print("\nGroup by (multi-column): with one column sum\n",df1)
```

```python
#Group by (multi-column): with multi column sum
df2=df.groupby(['A', 'B'], as_index=False).sum()
print("\nOriginal DataFrame\n",df)
print("\nGroup by (multi-column): with multi columns sum\n",df2)
```

```python
#combined groupby and aggregate function
df3=df.groupby(['A', 'B'], as_index=False)['C'].agg('sum')
print("\nOriginal DataFrame\n",df)
print("\nCombined Groupby and Aggregate function\n",df3)
```

```python
#combined groupby and aggregate function - separate column headings
df4 =(df.groupby(['A', 'B']).agg([('average', 'mean'), ('total', 'sum')]))
print("\nOriginal DataFrame\n",df)
print("\nCombined Groupby and Aggregate function-separate column headings\n",df4)
```

```python
df5=df.groupby(['A', 'B'], as_index=False).sum()
df6 = (df.groupby(['A', 'B']).agg(['sum']))
print("\nOriginal DataFrame\n",df)
print("\nGroupby-sum function\n",df5)
print("\nGroupby and Aggregate - sum function\n",df6)
```
