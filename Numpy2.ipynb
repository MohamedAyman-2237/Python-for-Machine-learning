```python
import numpy as np
arr=np.array([[1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
print("Original array:\n",arr)
```

```python
print("\nEvery other rows: \n",arr[:3:2])
```

```python
arr=np.array([1,2,3,4,5,6,7])
print("\nOriginal array:",arr)
print("\nReturns every other elements in the array : [::2]:",arr[::2])
```

```python

temp=arr[:2,:3]
print("\nArray with first 2 rows and 3 columns:\n",temp)
```

```python
cond=arr>2
temp=arr[cond]
print("\nElements greater than 2:\n",temp)
```

```python
import numpy as np
arr = np.array([[-1, 2, 0, 4], [4, -0.5, 6, 0], [2.6, 0, 7, 8], [3, -7, 4, 2.0]])
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print("Orginal arrays:\n",arr1,arr2)
print("\nJoined array:\n",arr)
```

```python
arr=np.hstack((arr1,arr2))
print("\nHorizontal joining:\n",arr)
```

```python
arr=np.vstack((arr1,arr2))
print("\nVertical joining:\n",arr)
```

```python
arr=np.dstack((arr1,arr2))
print("\nDepth joining:\n",arr)
```

```python
arr=np.array([1,2,3,4,5,6])
newarr=np.array_split(arr,3)
print("Orginal Array:\n",arr)
print("\nSplitted array:\n",newarr)
print(("\nSplitted array in another form:\n"))
print(newarr[0])
print(newarr[1])
print(newarr[2])
```
