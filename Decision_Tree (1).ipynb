## Aim

To implement Decision tree machine learning algorithm.

## Description

1. Import Decision tree classifier through sklearn
2. Provide the necessary dataset through CSV file
3. As per the trained dataset, decision tree can be obtained.

```python
import matplotlib.pyplot as plt
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv("d:\\decision_tree.csv")

print(df)
```

```python
d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)
```

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
```

```python
tree.plot_tree(dtree, feature_names=features)

plt.show()
```
