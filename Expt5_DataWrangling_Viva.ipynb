# Experiment 5 - Data Wrangling

**Q1. What is data wrangling in Python?**

Data wrangling (also called data munging) is the process of cleaning, transforming, and organizing raw data into a usable format. In Python, it is mainly done using the pandas library to reshape and prepare data for analysis.

---

**Q2. What are the main steps involved in data wrangling?**

The main steps include collecting raw data, cleaning missing/duplicate values, transforming data types, filtering irrelevant data, merging datasets, and reshaping data. These steps prepare the data for analysis or modeling.

---

**Q3. Which library is most commonly used for data wrangling in Python?**

Pandas is the most commonly used library for data wrangling in Python. It provides powerful tools like DataFrames, Series, groupby, merge, and pivot_table that make it easy to manipulate and transform data.

---

**Q4. What is the use of dropna() in Pandas?**

`dropna()` is used to remove rows or columns that contain missing (NaN) values from a DataFrame. By default it removes rows, but using `axis=1` removes columns. It helps clean incomplete data before analysis.

---

**Q5. What is the use of fillna() in Pandas?**

`fillna()` is used to fill missing (NaN) values in a DataFrame with a specified value or method. For example, `df.fillna(0)` fills all NaN values with 0, or `df.fillna(df.mean())` fills with column means.

---

**Q6. What is the purpose of the groupby() function in Pandas?**

`groupby()` is used to split data into groups based on one or more columns, apply an aggregation function to each group, and combine the results. It is useful for computing group-level statistics like sum, mean, or count.

---

**Q7. How do you rename columns in a DataFrame?**

You can rename columns using `df.rename(columns={'old_name': 'new_name'})`. You can also directly assign a new list of column names using `df.columns = ['col1', 'col2', ...]` if renaming all columns at once.

---

**Q8. What is the purpose of the merge() function in Pandas?**

`merge()` is used to combine two DataFrames based on a common column or index, similar to SQL JOIN operations. It supports inner, outer, left, and right joins using the `how` parameter.

---

**Q9. What is the difference between merge() and concat() in Pandas?**

`merge()` combines DataFrames based on common columns or keys (like SQL joins), while `concat()` simply stacks DataFrames either row-wise (axis=0) or column-wise (axis=1) without needing a common key.

---

**Q10. Why is data wrangling important in data analysis?**

Data wrangling ensures that data is in a clean, consistent, and structured format before analysis. Without proper wrangling, analysis results can be misleading. It is estimated that data scientists spend 80% of their time on this step.
