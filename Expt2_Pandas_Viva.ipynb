# Experiment 2 - Pandas Viva Questions

**Q1. What is pandas and why is it used?**

Pandas is a Python library used for data manipulation and analysis. It provides easy-to-use data structures like Series and DataFrame that allow efficient handling, cleaning, and analysis of structured data.

---

**Q2. What are the two main data structures in pandas?**

The two main data structures in pandas are Series (1-dimensional labeled array) and DataFrame (2-dimensional labeled table with rows and columns). Both support various data types and operations.

---

**Q3. How do you create a pandas Series from a Python list?**

You can create a Series using `pd.Series([10, 20, 30])`. This creates a one-dimensional labeled array. You can also provide custom index labels using the `index` parameter.

---

**Q4. How do you create a pandas DataFrame from a dictionary?**

You can create a DataFrame using `pd.DataFrame({'col1': [1,2], 'col2': [3,4]})`. Each key becomes a column name and its list of values forms the column data.

---

**Q5. How can you read a CSV file into a pandas DataFrame?**

You can read a CSV file using `pd.read_csv('filename.csv')`. This function reads the file and returns a DataFrame. You can also pass additional parameters like `sep`, `header`, and `index_col`.

---

**Q6. How do you display the first 5 rows of a DataFrame?**

You can use `df.head()` to display the first 5 rows of a DataFrame. You can also pass a number like `df.head(10)` to display the first 10 rows.

---

**Q7. What does the info() method do in a pandas DataFrame?**

The `info()` method displays a concise summary of the DataFrame including the number of entries, column names, non-null counts, and data types. It helps quickly understand the structure of the data.

---

**Q8. How do you select a specific column in a DataFrame?**

You can select a column using `df['column_name']` or `df.column_name`. This returns the column as a pandas Series. To select multiple columns, use `df[['col1', 'col2']]`.

---

**Q9. How do you filter rows in a DataFrame based on a condition?**

You can filter rows using boolean indexing: `df[df['column'] > value]`. For example, `df[df['Age'] > 18]` returns all rows where the Age column is greater than 18.

---

**Q10. How do you add a new column to a pandas DataFrame?**

You can add a new column by assigning values: `df['new_col'] = values`. For example, `df['Total'] = df['Maths'] + df['Science']` creates a new column named Total with the sum of two columns.
