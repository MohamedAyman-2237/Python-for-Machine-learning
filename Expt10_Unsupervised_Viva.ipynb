# Experiment 10 - Unsupervised Learning

**Q1. What is unsupervised learning?**

Unsupervised learning is a type of machine learning where the model is trained on data without labeled outputs. The model tries to find hidden patterns, structures, or groupings in the data on its own.

---

**Q2. Does unsupervised learning use labeled data?**

No, unsupervised learning does not use labeled data. The algorithm works with raw input data and discovers patterns or groupings without any predefined correct answers or output labels.

---

**Q3. What is clustering?**

Clustering is an unsupervised learning technique that groups data points into clusters based on their similarity. Points within the same cluster are more similar to each other than to points in other clusters. KMeans is a popular clustering algorithm.

---

**Q4. Give one example of unsupervised learning.**

Customer segmentation is an example of unsupervised learning, where customers are grouped based on purchasing behavior without predefined labels. Another example is anomaly detection in network traffic to find unusual patterns.

---

**Q5. What is grouping of data?**

Grouping of data means organizing data points into subsets (clusters) based on their similarities. In unsupervised learning, this is done automatically by algorithms like KMeans without human-defined class labels.

---

**Q6. What is the goal of unsupervised learning?**

The goal of unsupervised learning is to discover hidden patterns, structures, or groupings in unlabeled data. It aims to understand the underlying distribution of the data without any prior knowledge of the output.

---

**Q7. Name one unsupervised learning algorithm.**

KMeans Clustering is a popular unsupervised learning algorithm that partitions data into K clusters by minimizing the distance between data points and their cluster centroids. Other examples include DBSCAN and Hierarchical Clustering.

---

**Q8. What is pattern discovery?**

Pattern discovery is the process of finding recurring structures, relationships, or regularities in data without predefined labels. It is a key goal of unsupervised learning, helping uncover insights from raw data.

---

**Q9. What is the difference between supervised and unsupervised learning?**

Supervised learning uses labeled data and predicts a specific output (class or value), while unsupervised learning uses unlabeled data and discovers hidden patterns or groupings. Supervised learning has a target variable; unsupervised learning does not.

---

**Q10. Where is unsupervised learning used?**

Unsupervised learning is used in customer segmentation, anomaly detection, recommendation systems, image compression, and topic modeling in NLP. It is useful when labeled data is expensive or unavailable.
