# Experiment 8 - Time Series

**Q1. What is a time series?**

A time series is a sequence of data points recorded or measured at successive points in time, usually at uniform intervals. It is used to analyze how a variable changes over time, such as stock prices or temperature readings.

---

**Q2. Give one example of time series data.**

Daily temperature readings recorded over a year is an example of time series data. Other examples include monthly sales figures, hourly electricity consumption, and weekly stock prices — all indexed by time.

---

**Q3. What is the role of time in time series analysis?**

Time acts as the index in time series analysis, defining the order and spacing of observations. It allows detection of patterns like trends (long-term increase/decrease), seasonality (repeating patterns), and irregular fluctuations.

---

**Q4. What is a time index in Python?**

A time index in Python (using pandas) is a `DatetimeIndex` that labels rows of a DataFrame with timestamps. It enables time-based slicing, resampling, and rolling operations using `pd.date_range()` or `pd.to_datetime()`.

---

**Q5. Which Python library is commonly used for time series analysis?**

Pandas is the most commonly used library for time series analysis, providing `DatetimeIndex`, `resample()`, and `rolling()`. Libraries like Statsmodels and Prophet are also used for advanced forecasting.

---

**Q6. What is resampling in time series?**

Resampling is the process of changing the frequency of time series data. Downsampling reduces frequency (e.g., daily to monthly) and upsampling increases it (e.g., monthly to daily). In pandas, it is done using `df.resample('M').mean()`.

---

**Q7. What is the difference between daily and monthly data?**

Daily data has one observation per day, providing fine-grained detail. Monthly data aggregates values over a month, giving a broader trend view. Monthly data reduces noise but loses short-term fluctuations visible in daily data.

---

**Q8. What is a moving average?**

A moving average is the average of a fixed number of consecutive data points that moves forward as new data is added. It smooths out short-term fluctuations. In pandas: `df['col'].rolling(window=7).mean()` computes a 7-day moving average.

---

**Q9. Why do we handle missing values in time series?**

Missing values in time series can break continuity and cause errors in analysis and forecasting. They are handled using forward fill (`ffill`), backward fill (`bfill`), or interpolation to maintain a consistent time-indexed sequence.

---

**Q10. What is trend in time series data?**

Trend is the long-term upward or downward movement in time series data over an extended period. For example, a consistent rise in house prices over 10 years represents an upward trend. It is identified by smoothing or decomposition.
