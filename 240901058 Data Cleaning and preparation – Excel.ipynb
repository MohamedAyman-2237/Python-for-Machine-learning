Aim:

To perform data cleaning on an Excel dataset using Pandas by handling missing values,
removing unwanted spaces, modifying columns and rows, and extracting specific records.
Finally, to save the cleaned and updated dataset back into an Excel file.

Description:

Read the Excel file, do the data cleaning process and write the updated data set into excel file
1. Remove the white space using str.strip function
2. Fill the forward values for NaN using fillna(pad) method
3. Drop one particular column using drop function
4. Drop NaN rows using dropna function
5. Replace the values(s) using replace function
6. Extract the particular record based on the isin() function condition

```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original data:\n")
df=pd.read_excel("/content/II EEE marks.xlsx")
print(df)
```

```python
f=df['Student Name '].str.strip()
f.to_excel("/content/II EEE marks.xlsx")

print("\nReplace value:")
k=df.fillna(method='pad')
print(k)
print("\nUpdated list\n")
k.to_excel("/content/II EEE marks.xlsx")
```

```python
print("\nDrop one particular column and its values:\n")
k.drop(['Tamil'],axis=1,inplace=True)
print(k)
print("\nUpdated list..\n")
k.to_excel("/content/II EEE marks.xlsx")
```

```python
print("\nDrop NaN rows:\n")
df=pd.read_excel("/content/II EEE marks.xlsx")
print(df)
x=df.dropna()
print(x)
print("\nUpdated data list:\n")
x.to_excel("/content/II EEE marks.xlsx")
```

```python
print("\nReplace values:\n")
n=pd.read_excel( "/content/II EEE marks.xlsx")
print("\nOriginal data:\n")
print(n)
y=n.replace({89:88})
print("\nReplaced dataset:\n")
print(y)
print("\nUpdated dataset:\n")
y.to_excel("/content/II EEE marks.xlsx")
```

```python
df=pd.read_excel("/content/II EEE marks.xlsx")
print(df)
print("\nExtract using isin() function:\n")
new=df['Eng'].isin([63])
print(df[new])
```
