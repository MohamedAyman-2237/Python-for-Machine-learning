# Experiment 6 - Plotting and Visualization

**Q1. What is data visualization in Python?**

Data visualization is the graphical representation of data using charts, graphs, and plots. In Python, it helps in understanding patterns, trends, and relationships in data through visual means, making complex data easier to interpret.

---

**Q2. Which library is commonly used for plotting in Python?**

Matplotlib is the most commonly used library for plotting in Python. Other popular libraries include Seaborn (built on Matplotlib for statistical plots), Plotly (for interactive charts), and Pandas (which has built-in plotting).

---

**Q3. What is the purpose of Matplotlib?**

Matplotlib is a Python library used to create static, animated, and interactive visualizations. It provides functions to create line plots, bar charts, histograms, scatter plots, pie charts, and more for data analysis.

---

**Q4. What is the difference between plot() and scatter() in Matplotlib?**

`plot()` is used to draw line plots connecting data points, making it ideal for showing trends over time. `scatter()` displays individual data points as dots without connecting them, useful for showing relationships between two variables.

---

**Q5. What is a line plot?**

A line plot displays data points connected by straight lines, showing trends or changes over a continuous variable like time. It is created using `plt.plot(x, y)` and is commonly used in time series analysis.

---

**Q6. What is a bar chart and when is it used?**

A bar chart uses rectangular bars to represent data values for different categories. It is used to compare discrete categories or groups. It is created using `plt.bar(x, height)` for vertical bars.

---

**Q7. What is a histogram and what does it represent?**

A histogram displays the frequency distribution of a continuous numerical variable by dividing data into bins. It shows how data is distributed and is created using `plt.hist(data, bins=n)`.

---

**Q8. What is a scatter plot used for?**

A scatter plot is used to visualize the relationship or correlation between two numerical variables. Each point represents one observation. It is created using `plt.scatter(x, y)` and helps identify patterns or clusters.

---

**Q9. What is the use of xlabel() and ylabel() in Matplotlib?**

`plt.xlabel('label')` sets the label for the x-axis and `plt.ylabel('label')` sets the label for the y-axis. These labels help readers understand what each axis represents in the plot.

---

**Q10. What is the purpose of the title() function in Matplotlib?**

`plt.title('Title')` adds a title to the top of the plot. It provides a brief description of what the chart represents, making it easier for viewers to understand the purpose of the visualization.

---

**Q11. What is the use of legend() in Matplotlib?**

`plt.legend()` displays a legend on the plot that maps line colors or markers to their corresponding labels. It is useful when multiple datasets are plotted together, helping differentiate between them.

---

**Q12. What is the function used to display a plot in Matplotlib?**

`plt.show()` is used to display the plot on screen. It renders all the plot elements and opens the visualization window. In Jupyter notebooks, plots are often displayed inline without needing `plt.show()`.

---

**Q13. What is the difference between a bar chart and a histogram?**

A bar chart is used for categorical data where bars represent different categories with gaps between them. A histogram is used for continuous numerical data, where bars represent frequency of values within intervals (bins) with no gaps.

---

**Q14. What is the purpose of the subplot() function in Matplotlib?**

`plt.subplot(rows, cols, index)` allows multiple plots to be displayed in a single figure. It divides the figure into a grid and each call creates a plot in the specified position, useful for comparing multiple charts.

---

**Q15. What is the use of the figure() function in Matplotlib?**

`plt.figure(figsize=(width, height))` creates a new figure and allows you to set its size. It is used to control the overall dimensions and properties of the plot window before adding subplots or axes.

---

**Q16. What is the purpose of grid() in Matplotlib?**

`plt.grid(True)` adds a grid to the background of the plot. The grid helps readers accurately read values from the chart by providing reference lines across the plot area.

---

**Q17. What is a pie chart and when is it used?**

A pie chart is a circular chart divided into slices to show the proportion of each category relative to the whole. It is used when you want to show percentage distribution and is created using `plt.pie(values, labels=labels)`.

---

**Q18. Why is visualization important in data analysis?**

Visualization makes it easier to understand complex data, identify patterns, trends, and outliers at a glance. It helps communicate findings effectively to both technical and non-technical audiences, aiding better decision-making.
