```python
#1a
# Exam scores of 20 students
scores = [77, 67, 89, 53, 55, 90, 33, 71, 84, 49,
          67, 58, 100, 40, 76, 81, 53, 60, 47, 69]

# 1. Overall class performance
average_score = sum(scores) / len(scores)
highest_score = max(scores)
lowest_score = min(scores)

print("Class Performance:")
print("Average Score:", average_score)
print("Highest Score:", highest_score)
print("Lowest Score:", lowest_score)


# 2. Pass/Fail classification
print("\nPass/Fail Classification (Pass mark = 50):")
for i, score in enumerate(scores, start=1):
    if score >= 50:
        print(f"Student {i}: {score} - Pass")
    else:
        print(f"Student {i}: {score} - Fail")


# 3. Grade assignment
def assign_grade(score):
    if score >= 85:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"

print("\nGrades:")
for i, score in enumerate(scores, start=1):
    grade = assign_grade(score)
    print(f"Student {i}: Score = {score}, Grade = {grade}")
```

```python
#2a
import numpy as np

# Step count data for 30 days
steps = np.array([
6500, 7200, 8000, 5000, 9100, 10000, 7500,
8200, 6400, 7000, 9500, 11000, 6700, 7200,
7800, 8500, 9200, 6000, 7100, 8800, 9400,
10200, 7600, 6900, 8300, 9700, 10500, 7300,
8100, 8900
])

# 1. Key statistics
total_steps = np.sum(steps)
average_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print("Total steps:", total_steps)
print("Average steps per day:", average_steps)
print("Maximum steps:", max_steps)
print("Minimum steps:", min_steps)

# 2. Identify high activity days
threshold = 9000
high_activity_days = np.where(steps > threshold)

print("\nDays with high activity (steps >", threshold, "):")
print(high_activity_days[0] + 1)

# 3. Trend analysis
daily_change = np.diff(steps)

print("\nDaily change in steps:")
print(daily_change)
```

```python
#3a
import numpy as np

# Function for series impedance
def series_impedance(impedances):
    return sum(impedances)

# Function for parallel impedance
def parallel_impedance(impedances):
    inv_total = sum(1/z for z in impedances)
    return 1 / inv_total

# Number of components
n = int(input("Enter number of components: "))

impedances = []

# Input impedance values
print("Enter impedance values (real and imaginary parts):")
for i in range(n):
    real = float(input(f"Real part of Z{i+1}: "))
    imag = float(input(f"Imaginary part of Z{i+1}: "))
    z = complex(real, imag)
    impedances.append(z)

# Circuit type
circuit = input("Enter circuit type (series/parallel): ").lower()

if circuit == "series":
    Z_total = series_impedance(impedances)

elif circuit == "parallel":
    Z_total = parallel_impedance(impedances)

else:
    print("Invalid circuit type")
    Z_total = None

# Output result
if Z_total is not None:
    print("Total Impedance =", Z_total)
```

```python
#3b
import numpy as np

# Stock A price data (5 days × 3 features: Open, High, Low)
stock_A = np.array([
    [100, 105, 98],
    [102, 108, 101],
    [104, 110, 103],
    [103, 107, 100],
    [105, 111, 104]
])

# Stock B price data (5 days × 3 features: Open, High, Low)
stock_B = np.array([
    [98, 103, 95],
    [101, 106, 99],
    [103, 109, 100],
    [102, 105, 98],
    [104, 110, 101]
])

# 1. Element-wise sum
sum_matrix = stock_A + stock_B

# 2. Element-wise difference
difference_matrix = stock_A - stock_B

print("Stock A Matrix:")
print(stock_A)

print("\nStock B Matrix:")
print(stock_B)

print("\nElement-wise Sum of Matrices:")
print(sum_matrix)

print("\nElement-wise Difference (Stock A - Stock B):")
print(difference_matrix)
```

```python
import pandas as pd

# EEG dataset
d = {
    "Participant": ["P1","P1","P1","P2","P2","P2","P3","P3","P3"],
    "Electrode": ["E1","E2","E3","E1","E2","E3","E1","E2","E3"],
    "Brainwave": [12.5,15.2,14.1,13.3,16.5,15.8,11.9,14.7,13.8]
}

data = pd.DataFrame(d)

# Display data
print("EEG Data Sample:")
print(data.head())

# 1. Basic statistics
print("\nMean value:")
print(data["Brainwave"].mean())

print("\nVariance:")
print(data["Brainwave"].var())

print("\nPeak value:")
print(data["Brainwave"].max())

# 2. Statistics per participant
participant_stats = data.groupby("Participant")["Brainwave"].agg(['mean','var','max'])
print("\nStatistics per Participant:")
print(participant_stats)

# 3. Statistics per electrode
electrode_stats = data.groupby("Electrode")["Brainwave"].agg(['mean','var','max'])
print("\nStatistics per Electrode:")
print(electrode_stats)

# 4. Trend analysis (rolling mean)
data["Trend"] = data["Brainwave"].rolling(window=2).mean()

print("\nBrainwave Trend:")
print(data)
```

```python
#4b
import pandas as pd
import matplotlib.pyplot as plt

# Given inventory data
data = {
    'Product': ['Laptop','Smartphone','Tablet','Monitor','Keyboard'],
    'Price': [1200,800,300,150,50],
    'Quantity': [10,20,15,25,50]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Inventory Data:")
print(df)

# 1. Average price
print("\nAverage Price:", df['Price'].mean())

# 2. Maximum price
print("Maximum Price:", df['Price'].max())

# 3. Minimum price
print("Minimum Price:", df['Price'].min())

# 4. Total quantity
print("Total Quantity:", df['Quantity'].sum())

# 5. Total value column
df['Total Value'] = df['Price'] * df['Quantity']
print("\nInventory with Total Value:")
print(df)

# Plot graph
plt.bar(df['Product'], df['Price'])
plt.title("Product Price Comparison")
plt.xlabel("Product")
plt.ylabel("Price")
plt.show()
```
