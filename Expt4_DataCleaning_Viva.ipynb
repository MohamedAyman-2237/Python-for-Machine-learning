# Experiment 4 - Data Cleaning and Preparation

**Q1. What is meant by data cleaning?**

Data cleaning is the process of identifying and correcting errors, inconsistencies, and inaccuracies in a dataset. It involves handling missing values, removing duplicates, fixing formatting issues, and correcting wrong data.

---

**Q2. Why is data cleaning important before data analysis?**

Data cleaning is important because raw data often contains errors and inconsistencies that can lead to incorrect results. Clean data ensures that analysis and machine learning models produce accurate and reliable outcomes.

---

**Q3. What are missing values in a dataset?**

Missing values are entries in a dataset where no data is recorded for a particular field. In pandas, missing values are represented as `NaN` (Not a Number) and can result from data collection errors or incomplete records.

---

**Q4. How can missing values be handled in Python?**

Missing values can be handled using `df.dropna()` to remove rows with missing values, or `df.fillna(value)` to replace them with a specific value like mean, median, or mode. The choice depends on the dataset context.

---

**Q5. What is meant by data preparation?**

Data preparation involves transforming raw data into a clean and suitable format for analysis or modeling. It includes steps like cleaning, encoding categorical variables, scaling features, and splitting data into training and testing sets.

---

**Q6. What is duplicate data, and why should it be removed?**

Duplicate data refers to rows in a dataset that are identical or nearly identical. Duplicates can skew statistical analysis and model training, leading to biased results. They can be removed using `df.drop_duplicates()` in pandas.

---

**Q7. What is data inconsistency?**

Data inconsistency occurs when the same data is represented in different formats or with conflicting values across a dataset. For example, 'Male' and 'M' representing the same category. It must be standardized before analysis.

---

**Q8. How do you identify outliers in data?**

Outliers can be identified using statistical methods like the IQR (Interquartile Range) method or Z-score. Visualizations like box plots and scatter plots also help in spotting data points that are far from the rest.

---

**Q9. What is data normalization, and why is it needed?**

Data normalization is the process of scaling numerical data to a common range, typically 0 to 1. It is needed when features have different scales, ensuring that no single feature dominates the model during training.

---

**Q10. How does data cleaning improve results in engineering applications?**

Clean data leads to more accurate models, reliable predictions, and meaningful insights. In engineering, it prevents errors in system analysis, ensures sensor data is valid, and improves the performance of machine learning models.
