# Time Series 1

**Aim:**  
To implement and check Time Series functions in Python.

**Description:**  
1. Get today's, previous day and next day dates  
2. Display the data format from the raw data  
3. Replacing the day/month/year with the `replace` function  
4. Display day/month/year from the current date

### 1. Get current date and time using `datetime.now()`

```python
import datetime

a = datetime.datetime.now()
print("today-now function: ", a)
```

### 2. Get today's date using `datetime.today()`

```python
today = datetime.datetime.today()
print("today: ", today)
```

### 3. Get tomorrow's date (3 days ahead using `timedelta`)

```python
tomorrow = today + datetime.timedelta(days=3)
print("tomorrow:", tomorrow)
```

### 4. Get yesterday's date (2 days behind using `timedelta`)

```python
yesterday = today - datetime.timedelta(days=2)
print("yesterday: ", yesterday)
```

### 5. Display a specific date format using `datetime()`

```python
a = datetime.datetime.now()
b = datetime.datetime(2020, 6, 8, 23, 10, 25, 404040)
print("date format [2020,6,8,23,10,25,404040):", b)
```

### 6. Replace day in a date using `.replace()`

```python
from datetime import date

d = date(2024, 12, 31)
print("Replacing date", d, d.replace(day=15))
```

### 7. Display day and month using `.ctime()`

```python
t = date(2004, 4, 17)
d = date(2004, 4, 17).ctime()
print("Day and Month:", t, d)
```

### 8. Extract year, month, and day from today's date

```python
today = date.today()
print("Today:", today)
print("Current year:", today.year)
print("Current month:", today.month)
print("Current day:", today.day)
```
