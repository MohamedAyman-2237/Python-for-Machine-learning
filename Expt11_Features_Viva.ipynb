# Experiment 11 - Data Representation and Engineering Features

**Q1. What is a feature in machine learning?**

A feature is an individual measurable property or characteristic of the data used as input to a machine learning model. For example, in a house price prediction model, features include area, number of rooms, and location.

---

**Q2. What is feature engineering?**

Feature engineering is the process of using domain knowledge to create, transform, or select features from raw data to improve model performance. It involves operations like encoding, scaling, binning, and creating new derived features.

---

**Q3. Why is feature engineering important?**

Feature engineering improves model accuracy by providing better input representations. Good features help the model learn patterns more effectively, reduce training time, and often have a greater impact than choosing a complex algorithm.

---

**Q4. What is data representation?**

Data representation refers to how data is structured and encoded for use in machine learning models. Since models work with numerical data, raw data (like text or categories) must be converted into numerical form before training.

---

**Q5. What is numerical data?**

Numerical data consists of measurable quantities represented as numbers. It can be continuous (e.g., temperature, height) or discrete (e.g., number of rooms). Numerical data can be directly used in most machine learning algorithms.

---

**Q6. What is categorical data?**

Categorical data represents groups or categories with no inherent numerical meaning, such as color (Red, Blue), gender (Male, Female), or transmitter type (T1, T2). It must be encoded into numbers before being used in ML models.

---

**Q7. What is encoding in data representation?**

Encoding is the process of converting categorical data into numerical form so that machine learning algorithms can process it. Common encoding methods include Label Encoding (assigns integers) and One-Hot Encoding (creates binary columns).

---

**Q8. What is normalization?**

Normalization is the process of scaling numerical data to a range of 0 to 1. It is done using the formula: `(x - min) / (max - min)`. It ensures features with large values do not dominate those with smaller values during training.

---

**Q9. What is scaling of data?**

Scaling adjusts the range of feature values so they are on a comparable scale. Standardization (Z-score scaling) transforms data to have mean=0 and std=1. Scaling improves convergence speed and model performance in distance-based algorithms.

---

**Q10. Give one example of feature transformation.**

Binning (or discretization) is a feature transformation where continuous values are divided into discrete intervals. For example, converting age values into categories like Baby/Toddler, Child, Adult, and Elderly using `pd.cut()`.
