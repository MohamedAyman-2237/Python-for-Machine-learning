## Aim

To implement Linear Regression supervised machine learning algorithm.

## Description

1. Import stats for linear regression through scipy
2. Provide the necessary dataset through DataFrames
3. Finally we can obtain the linear regression output through matplotlib as graph

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))
```

```python
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
