```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
#seaborn package
Xpoints=np.array([0,20,40,60,80,120,200])
Ypoints=np.array([10,80,100,120,150,180,250])
plt.plot(Xpoints,Ypoints)
```

```python
sns.scatterplot(x=Xpoints, y=Ypoints)
plt.show()
```

```python
sns.barplot(x=Xpoints, y=Ypoints)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Data
X = np.array([1, 2, 6, 8, 12, 14, 20])
Y = np.array([3, 8, 1, 10, 12, 16, 18])
Z = np.array([5, 7, 9, 11, 13, 15, 17]) # You NEED a Z axis for 3D
# Create figure
fig = plt.figure(figsize=(12, 8))
# Create 3D axes
ax = plt.axes(projection='3d')
# 3D scatter plot
plt_3d = ax.scatter3D(X, Y, Z)
# Color bar
fig.colorbar(plt_3d)
# Show plot
plt.show()
```
