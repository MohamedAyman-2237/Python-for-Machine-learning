#### **Aim**:
To demonstrate data manipulation using Pandas by creating, merging, sorting and storing data in Excel files, and by reading and displaying CSV data. Also, to extract basic dataset information such as column names and the total number of records.

#### **Description:**
1. Create a DataFrame and store the data into specified Excel file

2. To read two Excel file data and merge through append function and store the merged data in to the new Excel file.

3. Using sort function, to sort and store the resultant data into a new Excel file

4. Read and display the CSV file

5. List the column headings and get the length of the table data

---
<br>

#### **Programs:**

```python
import pandas as pd
df=pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9]], index=['one','two', 'three'], columns = ['a', 'b', 'c'])
```

```python
#Taking the input from DataFrame and storing in the Excel file
print("Taking the input from DataFrame and storing in the Excel file")
df.to_excel("pandas_to_excel.xlsx", sheet_name = 'new_sheet_name')
```

```python
#Second DataFrame input to another Excel file
d = pd.DataFrame([[110, 210, 310], [12, 220, 320], [310, 320, 330]], index = ['four', 'five', 'six'], columns = ['a', 'b', 'c'])
d.to_excel('pandas_to_excel(1).xlsx', sheet_name = 'new_sheet_name')
```

```python
#Merging two Excel files input into third file
x=pd.read_excel("pandas_to_excel.xlsx")
y=pd.read_excel("pandas_to_excel(1).xlsx")
z=pd.concat([x, y])
z.to_excel('pandas_to_excel3.xlsx')
```

```python
#Sorting the column values
df=z.sort_values(['a'])
print(df)
df.to_excel('pandas_to_excel(4).xlsx')
```

```python
df=pd.read_excel('pandas_to_excel(4).xlsx')
print(df)
print(list(df))
print(format(len(df)))
```
